* **seasons:** Number of seasons (applicable only for TV shows).

* **imdb_score:** Rating of the title on IMDb.
* **imdb_votes:** Total number of votes received on IMDb.
* **tmdb_popularity:** Popularity score from TMDB database.
* **tmdb_score:** Rating score from TMDB.

**Credits Dataset Variables**

* **person_id:** Unique identifier for a person involved in the title.
* **id:** Identifier linking the person to a specific title.
* **name:** Name of the actor or director.
* **character_name:** Name of the character played by the actor.
